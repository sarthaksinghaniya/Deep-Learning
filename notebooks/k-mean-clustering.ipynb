```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Numerical data: Height and Weight
data_num = {
    'Height': [150, 160, 170, 180, 190, 200, 210],
    'Weight': [50, 60, 70, 80, 90, 100, 110]
}
df_num = pd.DataFrame(data_num)
print("Original Data:")
print(df_num)
```

```python
# Apply K-Means
kmeans = KMeans(n_clusters=2, random_state=0)
df_num['Cluster'] = kmeans.fit_predict(df_num)

# Output Result
print("=== K-MEANS Clustering Result ===")
print(df_num)
```

```python
# Visualize
plt.figure(figsize=(10, 6))
plt.scatter(df_num['Height'], df_num['Weight'], c=df_num['Cluster'], cmap='viridis', s=100, alpha=0.8)
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('K-Means Clustering (Height vs Weight)')
plt.grid(True)
plt.colorbar(label='Cluster')
plt.show()
```

```python
# Display cluster centers
print("Cluster Centers:")
for i, center in enumerate(kmeans.cluster_centers_):
    print(f"Cluster {i}: Height={center[0]:.2f}, Weight={center[1]:.2f}")
```

```python
# Find optimal number of clusters using Elbow Method
inertias = []
k_range = range(1, 6)

for k in k_range:
    kmeans_test = KMeans(n_clusters=k, random_state=0)
    kmeans_test.fit(df_num[['Height', 'Weight']])
    inertias.append(kmeans_test.inertia_)

# Plot Elbow Method
plt.figure(figsize=(10, 6))
plt.plot(k_range, inertias, 'bo-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.grid(True)
plt.show()
```
