```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8-whitegrid')
plt.rc('figure', autolayout=True)
plt.rc('axes', labelweight='bold', labelsize='large',
       titleweight='bold', titlesize=18, titlepad=10)

# Note: learntools is not available in this environment
# from learntools.core import binder
# binder.bind(globals())
# from learntools.deep_learning_intro.ex2 import *
```

```python
concrete = pd.read_csv('concrete.csv')
concrete.head()
```

```python
X = concrete.drop('CompressiveStrength', axis=1)
y = concrete['CompressiveStrength']
```

```python
# YOUR CODE HERE
input_shape = [X.shape[1]]

# Check your answer
# Note: q_1.check() is not available without learntools
# q_1.check()
print(f"Input shape: {input_shape}")
```

```python
# YOUR CODE HERE
model = keras.Sequential([
    layers.Dense(512, activation='relu', input_shape=input_shape),
    layers.Dense(512, activation='relu'),
    layers.Dense(512, activation='relu'),
    layers.Dense(1),
])

# Check your answer
# Note: q_2.check() is not available without learntools
# q_2.check()
print("Model created successfully")
model.summary()
```

```python
# YOUR CODE HERE
model = keras.Sequential([
    layers.Dense(32, input_shape=[8]),
    layers.Activation('relu'),
    layers.Dense(32),
    layers.Activation('relu'),
    layers.Dense(1),
])

# Check your answer
# Note: q_3.check() is not available without learntools
# q_3.check()
print("Alternative model created successfully")
model.summary()
```

```python
model.compile(
    optimizer='adam',
    loss='mae'
)
```

```python
history = model.fit(
    X, y,
    validation_split=0.2,
    batch_size=32,
    epochs=200,
    verbose=0
)
```

```python
history_df = pd.DataFrame(history.history)
history_df.plot()
plt.show()
```
